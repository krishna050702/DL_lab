Implement autoencoder for both binary and real valued input. clearly show the loss after each iteration <br>
number of inputs = 5 <br>
hidden layer = 4 <br>
binary cross entropy loss function and sigmoid activation function<br> 
input = 1 0 1 1 1 <br>

# Binary input

```python
import numpy as np
def sigmoid(x):
    return 1.0 / (1.0 + np.exp(-x))

def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

def binary_cross_entropy(y_true, y_pred):
    return -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred)) 

def binary_cross_entropy_derivative(y_true, y_pred):
    return -(y_true / y_pred) + (1 - y_true) / (1 - y_pred)

# input 
X = np.array([[1, 0, 1, 1, 1]])


input_size = X.shape[1]
hidden_size = 4
output_size = input_size

np.random.seed(0)
W1 = np.random.rand(input_size, hidden_size)
W2 = np.random.rand(hidden_size, output_size)

print(W1)
print(W2)

#  number of epochs and learning rate
epochs = 1000
learning_rate = 0.01


for i in range(epochs):
    # forward pass
    hidden_layer = sigmoid(np.dot(X, W1))
    output_layer = sigmoid(np.dot(hidden_layer, W2))
    # loss 
    loss = binary_cross_entropy(X, output_layer)
    print(f"Epoch {i}, Loss: {loss}")
    # backpropagation
    d_output = binary_cross_entropy_derivative(X, output_layer) * sigmoid_derivative(output_layer)
    d_hidden = np.dot(d_output, W2.T) * sigmoid_derivative(hidden_layer)
    # update weights
    W2 -= learning_rate * np.dot(hidden_layer.T, d_output)
    W1 -= learning_rate * np.dot(X.T, d_hidden)
# output 
reconstructed = sigmoid(np.dot(sigmoid(np.dot(X, W1)), W2))
print()
print("W1: \n",W1)
print("W2: \n",W2)

print("\n\nActual input:", X)
print("Reconstructed input:", reconstructed)
```

# Real valued Input

```python
import numpy as np

def linear(x):
    return x

def linear_derivative(x):
    return np.ones_like(x)

def mse(y_true, y_pred):
    return np.mean(np.square(y_true - y_pred))

def mse_derivative(y_true, y_pred):
    return 2 * (y_pred - y_true) / len(y_true)

# input 
X = np.array([[1.7, 2.7, 3.1, 1.9, 3.1]])

# shape of input, hidden and output layer
input_size = X.shape[1]
hidden_size = 4
output_size = input_size

# He initialization of weights
np.random.seed(0)
W1 = np.random.randn(input_size, hidden_size)
W2 = np.random.randn(hidden_size, output_size) 

# print initial weights
print(W1)
print(W2)

#  number of epochs and learning rate
epochs = 500
learning_rate = 0.001

for i in range(epochs):
    # forward pass
    hidden_layer = linear(np.dot(X, W1))
    output_layer = linear(np.dot(hidden_layer, W2))
    
    # loss 
    loss = mse(X, output_layer)
    print(f"Epoch {i}, Loss: {loss}")
    
    # backpropagation
    d_output = mse_derivative(X, output_layer) * linear_derivative(output_layer)
    d_hidden = np.dot(d_output, W2.T) * linear_derivative(hidden_layer)
    
    # update weights
    W2 -= learning_rate * np.dot(hidden_layer.T, d_output)
    W1 -= learning_rate * np.dot(X.T, d_hidden)

# output 
reconstructed = linear(np.dot(linear(np.dot(X, W1)), W2))
print()
print("W1: \n",W1)
print("W2: \n",W2)

print("\n\nActual input:", X)
print("Reconstructed input:", reconstructed)
```
