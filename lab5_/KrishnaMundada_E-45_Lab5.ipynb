```Write a program to implement backkpropagation algorithm on iris dataset.```
<hr>


````python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Load the iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Standardize the features
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Define the neural network architecture
input_size = X_train.shape[1]
hidden_size = 5
output_size = len(np.unique(y_train))

# Initialize the weights and biases
W1 = np.random.randn(input_size, hidden_size) * 0.01
b1 = np.zeros((1, hidden_size))
W2 = np.random.randn(hidden_size, output_size) * 0.01
b2 = np.zeros((1, output_size))

# Define the sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Train the neural network using backpropagation
epochs = 10000
learning_rate = 0.1

for i in range(epochs):
    # Forward propagation
    z1 = np.dot(X_train, W1) + b1
    a1 = sigmoid(z1)
    z2 = np.dot(a1, W2) + b2
    y_pred = np.argmax(z2, axis=1)
    
    # Compute the loss
    loss = np.sum(np.square(y_pred - y_train))
    
    # Backward propagation
    m = y_train.shape[0]
    dZ2 = np.zeros_like(z2)
    dZ2[range(m), y_train] = 1
    dZ2 -= 1
    dZ2 += y_pred.reshape(-1,1)
    dW2 = np.dot(a1.T, dZ2)
    db2 = np.sum(dZ2, axis=0, keepdims=True)
    dA1 = np.dot(dZ2, W2.T)
    dZ1 = dA1 * sigmoid(z1) * (1 - sigmoid(z1))
    dW1 = np.dot(X_train.T, dZ1)
    db1 = np.sum(dZ1, axis=0)
    
    # Update the weights and biases
    W1 -= learning_rate * dW1
    b1 -= learning_rate * db1
    W2 -= learning_rate * dW2
    b2 -= learning_rate * db2
    
    # Print the loss after every 1000 epochs
    if i % 1000 == 0:
        print("Epoch:", i, "Loss:", loss)

# Evaluate the trained model on the test set
z1 = np.dot(X_test, W1) + b1
a1 = sigmoid(z1)
z2 = np.dot(a1, W2) + b2
y_pred = np.argmax(z2, axis=1)
accuracy = np.mean(y_pred == y_test)
print("Test Accuracy:", accuracy)
````
